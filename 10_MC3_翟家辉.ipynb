# 一 · 中心极限定理介绍和证明

# > *`介绍 `*

> <font size=4 >**中心极限定理**是指概率论中讨论随机变量序列部分和分布渐近于正态分布的一类定理。这组定理是**数理统计学**和**误差分析**的理论基础，指出了大量随机变量近似服从**正态分布**的条件。在自然界与生产中，一些现象受到许多相互独立的随机因素的影响，如果每个因素所产生的影响都很微小时，总的影响可以看作是服从**正态分布**的。

>### 设随机变量 $X_1,X_2,\cdots,X_n,\cdots$ 相互独立且服从同一分布，该分布存在有限的期望和方差 $E(X_i)=\mu, Var(X_i)=\sigma^2, (i=1,2,\cdots)$ 。令
  $$Y_n = \frac{\sum_{k=1}^n {X_k-n\mu}}{\sqrt{n}\sigma}$$ 
>### 则 $n$ 趋于无穷大时，$Y_n$ 的分布趋向于标准正态分布 $N(0,1)$ 。即
$$\lim_{n \to +\infty}{P(Y_n<x)} = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{x} {e^{-\frac{t^2}{2}}}{\rm d}x$$
>### 据此得出如下两个推论：
>### 1.均值 $\overline x = \frac{\sum_{k=1}^n {X_k}}{n} = Y_n\frac{\sigma}{\sqrt{n}}+\mu$ 的分布趋向于正态分布 $N(\mu,\frac{\sigma^2}{n})$ 。
>### 2.n项和 $\sum_{k=1}^n {X_k} = \frac{\sum_{k=1}^n {X_k}}{n} = \sqrt{n}Y_n\sigma+n\mu$ 的分布趋向于正态分布 $N(n\mu,n\sigma^2)$ 。
>### 
### 对于独立同分布的随机变量 $X_1,X_2,\cdots,X_n,\cdots$ ，无论其服从何种分布，只要其期望和方差存在，当$n$充分大时，平均数 $\overline x$ 就近似服从正态分布 $N(\mu,\frac{\sigma^2}{n})$ 。因此可利用正态分布近似估计满足定理条件任意样本平均数落在某一区间概率。

# > *`证明 `*

>## `特征函数`
>### 设随机变量 $X$ 的概率密度函数为 $f(x)$ ，其特征函数为 $\varphi(t)=\int_{-\infty}^{\infty}{e^{itx}f(x){\rm d}x}$。
>### 若$Y=X_{1}+X_{2}$，$X_{1},X_{2}$相互独立，
>### $$\varphi_{y}(t)=\varphi_{x_{1}+x_{2}}(t)=∬_{-\infty}^{\infty}{e^{it(x_{1}+x_{2})}\cdot f(x_{1})\cdot g(x_{2})\cdot dx_{1}dx_{2}}=\int_{-\infty}^{\infty}{e^{itx_{1}}f(x_{1})dx_{1}}\cdot\int_{-\infty}^{\infty}{e^{itx_{2}}g(x_{2})dx_{2}}=\varphi_{x_{1}}(t)\cdot\varphi_{x_{2}}(t)$$
>### 若$Y=aX+b$,
>### $$\varphi_{y}(t)=\varphi_{ax+b}(t)=\int_{-\infty}^{\infty}e^{it(ax+b)}f(x)dx=e^{itb}\int_{-\infty}^{\infty}e^{i(at)x}f(x)dx=e^{itb}\cdot\varphi_{x}(at)$$
>### 若 $X\sim N(0,1)$ ，则其概率密度函数为 $f(x)=\frac{1}{\sqrt{2\pi}}e^{-\frac{x^{2}}{2}}$，特征函数为：
>### $$\varphi(t)=\int_{-\infty}^{\infty}e^{itx}\cdot\frac{1}{\sqrt{2\pi}}e^{-\frac{x^{2}}{2}}\cdot dx=e^{-\frac{t^2}{2}}\int_{-\infty}^{\infty}\frac{1}{\sqrt{2\pi}}e^{-\frac{{(x-it)}^{2}}{2}}d(x-it)=e^{-\frac{t^2}{2}}$$


>## `中心极限定理`
>### 设 $X_{1},X_{2},⋯,X_{n}$ 独立同分布，且$E(X)=\mu,D(X)=\sigma^2,S=\sum_{i=1}^{n}{X_i},\bar{X}=\frac{1}{n}S,Y=\frac{\bar{X}-\mu}{\sigma/\sqrt{n}}$
>### 由上式可知，$\varphi_{s}(t)=\left[ \varphi_x (t)\right]^n $，$\varphi_{\bar{x}}(t)=\varphi_{s}(\frac{t}{n})=\left[ \varphi_x (\frac{t}{n})\right]^n $
>### $\varphi_{y}(t)=e^{i(-\frac{\sqrt{n}}{\sigma}\mu)t}\cdot\varphi_{\bar{x}}(\frac{\sqrt{n}}{\sigma}t)=e^{i(-\frac{\sqrt{n}}{\sigma}\mu)t}\cdot\left[ \varphi_x (\frac{t}{\sigma \sqrt{n}})\right]^n$
>### 取对数得 $\ln\varphi_{y}(t)=\ln \left\{ e^{i(-\frac{\sqrt{n}}{\sigma}\mu)t}\cdot\left[ \varphi_x (\frac{t}{\sigma \sqrt{n}})\right]^n \right\}=-i\frac{\sqrt{n}}{\sigma}\mu t+n\ln \left[ \varphi_x (\frac{t}{\sigma \sqrt{n}}) \right]= \frac{-i\mu\frac{ t}{\sigma \sqrt{n}}+\ln \left[ \varphi_x (\frac{t}{\sigma \sqrt{n}}) \right]}{\frac{1}{n}}$
>### $\varphi_{x}(0)=\int_{-\infty}^{\infty}f(x)dx=1$，
>### $\varphi_{x}'(0)=\int_{-\infty}^{\infty}ixf(x)dx=i\mu$，
>### $\varphi_{x}''(0)=\int_{-\infty}^{\infty}-x^2f(x)dx=-E(X^2)=-\mu^2-\sigma^2$
>### 令 $p=\frac{t}{\sigma \sqrt{n}}$，
>## $\lim_{n\rightarrow\infty}\ln\varphi_{y}(t)=\lim_{n\rightarrow \infty}\frac{-i\mu\frac{ t}{\sigma \sqrt{n}}+\ln \left[ \varphi_x (\frac{t}{\sigma \sqrt{n}}) \right]}{\frac{1}{n}}=\frac{t^2}{\sigma^2}\lim_{p\rightarrow0 }\frac{-i\mu p+\ln \left[ \varphi_x (p) \right]}{p^2}=\frac{t^2}{\sigma^2}\lim_{p\rightarrow0 }\frac{-i\mu+\frac{1}{\varphi_x (p)}\cdot \varphi_x'(p)}{2p}=\frac{t^2}{\sigma^2}\lim_{p\rightarrow0 }\frac{\varphi_x''(p)\cdot\varphi_x(p)-\varphi_x'(p)\cdot \varphi_x'(p)}{2\left[ \varphi_x (p) \right]^2}=\frac{t^2}{\sigma^2}\cdot\frac{\varphi_x''(0)\cdot\varphi_x(0)-\varphi_x'(0)\cdot \varphi_x'(0)}{2\left[ \varphi_x (0) \right]^2}=\frac{t^2}{\sigma^2}\cdot\frac{(-\mu^2-\sigma^2)\cdot1-i\mu\cdot i\mu}{2\cdot1}=-\frac{t^2}{2}$
>## 即 $\lim_{n\rightarrow\infty}\varphi_{y}(t)=e^{-\frac{t^2}{2}}$, $Y\sim N(0,1)$

# 二 · 中心极限定理的数值验证

```python
#常用库及命令

import numpy as np

import matplotlib.pyplot as plt

import ipywidgets as widgets

from matplotlib.widgets import SpanSelector

import matplotlib.gridspec as gridspec

import random

import scipy as sc

from scipy.optimize import curve_fit

%matplotlib notebook

plt.rcParams['font.sans-serif'] = ['SimHei']
# 显示中文
plt.rcParams['axes.unicode_minus'] = False
# 显示负号
```

# > *`两点分布 `*

```python
def func(x,u,c,a):
    return(a*np.e**(-(x-u)**2/2/c)/np.sqrt(2*np.pi*c))

def fenbu(N,M,f,hhh):
    s=np.zeros(M)
    random.seed()
    for i in range(M):
        for ii in range(N):
            s[i]=s[i]+random.choice([-1,1])
    a=min(s)
    b=max(s)
    zj=(b-a)/f
    xs=np.linspace(-0.5*(b-a),0.5*(b-a),f)
    plt.figure(figsize=(15,8))
    n,bins,patches=plt.hist(s,xs,density=(hhh=="频率分布直方图"),edgecolor="black",label="数据值")
# 数据筛选
    x=[]
    y=[]
    
    for i in range(f-1):
        if n[i]!=0:
            x.append(bins[i])
            y.append(n[i])
    popt, pcov = curve_fit(func,x,y)
    print(popt,zj)
    print("\n")
    print("均值",np.mean(s),"方差",np.var(s))
    n=len(x)
    plt.plot(x,y,"c",linewidth=1.5)
    u=int(popt[0]*100)/100
    c=int(popt[1]*100)/100
    A=int(popt[2]*100)/100
    plt.plot(xs,func(xs,popt[0],popt[1],popt[2]),linewidth=3,label="拟合曲线")
    plt.title("两点分布",size=25)
    plt.xlabel("分布值",size=15)
    plt.ylabel('频率/组距 or 频数',size=15)
    plt.yticks(size=20,weight='bold')
    plt.xticks(size=18,weight='bold')
    plt.legend(loc='upper right',fontsize=20)
    

widgets.interact(
    fenbu,
    N=widgets.IntSlider(min=100, max=1000, step=100, value=300,description='N(行走次数):'),
    M=widgets.IntSlider(min=100, max=10000, step=100, value=1000,description='M(迭代点数):'),
    f=widgets.IntSlider(min=10, max=1000, step=10, value=200,description='分组数:'),
    hhh=widgets.RadioButtons(options=['频率分布直方图','频数分布直方图'],description='演化图类型:',disabled=False),
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from matplotlib.widgets import SpanSelector
import matplotlib.gridspec as gridspec
import random
import scipy.stats
from scipy.optimize import curve_fit

plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示中文
plt.rcParams['axes.unicode_minus'] = False  # 显示负号
```

# > *`一阶原点矩 ——均值`*

```python
def LM(r,初值,迭代次数):
    x0=初值
    N=迭代次数
    x=np.arange(0,N,1)
    y=[x0]
    for i in range(N):
        s=r*y[i]*(1-y[i])
        y.append(s)
    return y

def LMjv(N0,M,r):
    s=np.zeros(M)
    random.seed()

    for i in range(M):
        y=LM(r,random.uniform(0,1),N0)
        for ii in range(N0):
            s[i]=s[i]+random.choice(y)
            
    a=np.mean(s)
#     b=np.std(s)
#     c=scipy..stats.skew(s)
#     d=scipy..stats.kurtosis(s)
    return a

def LM1(N,M,r):
    n=np.arange(10,N,10)
    S=[]
    for i in n:
        S.append(LMjv(i,M,r))
    plt.figure(figsize=(15,8))    
    plt.plot(n,S,label="计算值")
    plt.title("L-M分布的均值",size=25)
    plt.xlabel("N",size=15)
    plt.ylabel('均值',size=15)
    plt.yticks(size=20,weight='bold')
    plt.xticks(size=18,weight='bold')
    plt.legend(loc='upper right',fontsize=20)
    

widgets.interact(
    LM1,
    N=widgets.IntSlider(min=1, max=1000, step=10, value=100,description='N(行走次数):'),
    M=widgets.IntSlider(min=100, max=10000, step=100, value=3000,description='M(迭代点数):'),
    r=[3.5699340,3.7,3.75,3.8,4],
)
```

```python
def LM(r,初值,迭代次数):
    x0=初值
    N=迭代次数
    x=np.arange(0,N,1)
    y=[x0]
    for i in range(N):
        s=r*y[i]*(1-y[i])
        y.append(s)
    return y

def LMjv(N0,M,r):
    s=np.zeros(M)
    random.seed()

    for i in range(M):
        y=LM(r,random.uniform(0,1),N0)
        for ii in range(N0):
            s[i]=s[i]+random.choice(y)
            
#     a=np.mean(s)
#     b=np.std(s)
#     c=scipy.skew(s)
    d=scipy.stats.kurtosis(s)
    return d

def LM1(N,M,r):
    n=np.arange(10,N,10)
    S=[]
    for i in n:
        S.append(LMjv(i,M,r))
    plt.figure(figsize=(15,8))    
    plt.plot(n,S,label="计算值")
    plt.title("L-M分布的峰度",size=25)
    plt.xlabel("N",size=15)
    plt.ylabel('峰度',size=15)
    plt.yticks(size=20,weight='bold')
    plt.xticks(size=18,weight='bold')
    plt.legend(loc='upper right',fontsize=20)
    

widgets.interact(
    LM1,
    N=widgets.IntSlider(min=1, max=1000, step=10, value=100,description='N(行走次数):'),
    M=widgets.IntSlider(min=100, max=10000, step=100, value=1000,description='M(迭代点数):'),
    r=[3.5699340,3.7,3.75,3.8,4],
)
```
