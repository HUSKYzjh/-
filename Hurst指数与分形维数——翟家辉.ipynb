# 【 一 】 文献背景

<font size=3.5>赫斯特指数描述为可以用来衡量时间序列的长期记忆的指标。它与时间序列的自相关性以及随着值对之间的滞后增加而降低的速率有关。赫斯特指数最初是在水文学中发展起来的，用于确定尼罗河挥发性降雨和干旱条件下的最佳水坝规模的实际问题，已被发现在市场营销、医学、工程和许多其他领域非常有用。赫斯特等人(1965)根据经验发现，自然界中的许多数据集满足幂律关系$(R/S)_n=(n/2)^H$。 $Bassingthwaighte$ 和 $Raymond$ 对分数阶高斯噪声和分数阶布朗步行进行了广泛的R/S分析。在研究过程中易与Hausdorff测度混淆，但Hausdorff测度Ha的值与Hurst指数Hu的值不严格相同,这可借助尼罗河流水说明：前者是指洪水事件的统计，后者是指流水总量和的统计。

# 【 二 】 定义和性质介绍

## 1. *分数布朗运动（Fractional Brownian Motion）*

<font size=3.5>分数布朗运动通常用于模拟复杂系统中的随机变化，如金融市场、气象学、地质学等。分数布朗运动的一个重要特征是其相关性，即数据间的相关性随时间的变化而变化。


>### a. 布朗运动的定义及性质

<font size=3.5>**布朗运动**是指一个连续随机过程。其最初是用来描述是悬浮在介质中的粒子的随机运动。这项运动以植物学家**罗伯特·布朗**（Robert Brown）的名字命名，他在1827年首次通过显微镜观察浸入水中的植物花粉而描述了这种现象。$\{B(t),t\geq0\}$为一个定义在非负实数（时域）$t$ 上的连续随机过程，若其为一维标准布朗运动（即**维纳过程**），则：

<font size=3.5>1. $ B(0)=0 $。即标准布朗运动在 $ t = 0 $ 时的位置为 $0$ 。

    
<font size=3.5>2. 对于所有的 $ 0\lt s\lt t $，增量 $\textit{B(t)–B(s)}$ 符合均值为 0，方差为 $t-s$ 的正态分布。即在任何有限时间区间 ${\Delta t}$ 内，布朗运动的变化满足均值为 $0$ 方差为 ${\Delta t}$ 的正态分布 $\mathcal{N}(0,\text{}\Delta t)$ ，其方差随时间区间的长度而线性增加。

    
<font size=3.5>3. 对于不重叠的区间 $\left[s_{i},{t}_{i}\right]$ ，随机变量 ${B}(t_i)-{B}(s_i)$ 之间是相互独立的。布朗运动在任何一个时间区间内的变化与其他与之不重叠的时间区间内的变化均无关。由该性质可知，布朗运动是一个马尔科夫过程，即该过程在任意 $t$ 时刻之后的位置仅和 $t$ 时刻的位置有关，与 $t$ 之前的历史轨迹无关。


>### b. 分数布朗运动的特性

<font size=3.5>**分数布朗运动**（fBM）又称**分形布朗运动**，表示一个数字在随机漫步过程中的变化，与布朗运动不同的是，分数布朗运动对时间相关度可以通过参数 $H$ 来控制。fBM 定义为一个在时域上的连续随机过程 $B_{H}(t)$ ,它满足以下两条特征：

<font size=3.5>1. 对于任何 $t$ 和 $\Delta t>0$ ，$\begin{aligned}B_H(t+\Delta t)\text{-}B_H(t)\end{aligned}$ 的期望为 $0$ ，即 fBM 的增量的期望为 0。
    
<font size=3.5>2. 对于不同时刻 $t$ 和 $s$ ，它们的协方差函数为 $\text{E}[B_H(t)B_H(s)]=\frac{1}{2}\Big(|t|^{2H}+|s|^{2H}-|t-s|^{2H}\Big)$。与经典布朗运动不同，fBm 的增量不必是独立的。
    
<font size=3.5>$H$ 是属于 $(0,1)$ 的实数，称为与分数布朗运动相关的**赫斯特指数**(Hurst Exponent)。Hurst 指数描述了由此产生的运动的粗糙度，其值越高，运动越平滑。分数布朗运动是非平稳的,但其增量是平稳的.对于分数布朗运动的一阶差分序列,又称为**分数高斯噪声**。计算 Hurst指数可以来判断简单分形时间序列是否具有长记忆性。关于Hurst指数的详细内容我们将在下节说明。


>### c. 模拟生成 fBM 的几种方法

<font size=3.5>**随机游走法**：该方法生成fBM的基本思想是把fBM看作一组随机游走的数据。从初始点出发，以固定概率向左或向右移动，生成的数据可以看作是一个随机过程。这种方法实现简单，但生成的数据精确度不高。

<font size=3.5>**卷积法**：卷积法是一种使用卷积的方式生成fBM的方法。该方法的基本思想是：先生成一组长度为N的随机数，然后对这组数据进行卷积操作，得到生成的fBM数据。这种方法生成的数据精确度较高，但计算量较大，实现比较复杂。

<font size=3.5>**周期性采样法**：周期性采样法是一种利用周期性采样的方式生成fBM的方法。该方法的基本思想是：在满足周期性的前提下对数据进行采样，然后对采样得到的数据进行合并，得到生成的fBM数据。这种方法生成的数据精确度较高，处理大数据集时效率较高。

<font size=3.5>**高斯过程生成法**：高斯过程生成法是一种使用高斯过程的方式生成fBM的方法。该方法的基本思想是：生成一组具有特定协方差矩阵的高斯随机数，然后将高斯随机数代入特定的函数中，得到生成的fBM数据。这种方法生成的数据精确度较高，处理大数据集时效率较高。

<font size=3.5>**生成函数法**：生成函数法是一种利用生成函数的方式生成fBM的方法。该方法的基本思想是：生成一组具有特定分布的随机数，然后代入特定的生成函数中，得到生成的fBM数据。这种方法生成的数据精确度较高，实现相对简单。

<font size=3.5>这些方法中，每种方法的适用范围和优劣势各不相同，在本文中我们采用高斯过程生成法下的**随机 Weierstrass 函数**生成 fBM 。

>### d. 随机魏尔斯特拉斯函数生成 fBM

<font size=3.5>**Weierstrass函数**是一类处处连续而处处不可导的实值函数。给定参数 $b>1$ 和 $0<H<1$，定义函数 $W(t)$ 为：

<font size=3.5>$$W\bigl(t\bigr)=\sum_{k=1}^{\infty}c_{k}b^{-H k}\sin\bigl(b^{k}t+d_{k}\bigr)$$

<font size=3.5>其中 $c_k$ 属于均值为 0 、标准差为 1 的正态分布，$d_k$ 在 $0≤d_k<2π$ 的区间内均匀分布。利用 Weierstrass 函数即可生成具有特定 $Hurst$ 指数的 fBm 数据。

```python
def weierstrass_function(x,Hurst_exponent,b):
    """
    计算随机 Weierstrass 函数，生成fBM数据
    
    参数：
    x - 时间序列
    Hurst_exponent - Hurst 指数
    b - Weierstrass 函数的类周期控制参量
    
    返回：
    对应x的Weierstrass 函数值
    
    """
    s = 0
    random.seed()
    #取前100项
    c=np.random.standard_normal(1000)
    d=np.random.uniform(0,2*np.pi,size=1000)
    for k in range(1000):
        s += c[k] * np.power(b,-(k+1) * Hurst_exponent) * np.sin(  np.power(b,k+1) * x + d[k])
    return s
```

## 2. *赫斯特指数（Hurst Exponent）*

<font size=3.5>**Hurst 指数** $H$ 被用于刻画数据长记忆性；它常被用来测量一个时间序列的波动范围如何随时间跨度变化而变化。1951年，英国水文学家赫斯特（Harold Edwin Hurst）在研究尼罗河水位变化时发现了时间序列中存在的**长记忆性**，即时间序列当前或过去的取值以远超随机扰动所能达到的程度影响该时间序列在未来的取值。如果一个平稳时间序列的自相关函数 $p(k)$ 的衰减速度服从**幂律衰减**（即比指数衰减慢），那么这个时间序列就具备长记忆性。记忆性体现在自相关函数的非独立性上，而“长”体现在衰减的速度慢。

>### a. Hurst 指数的来源

<font size=3.5>哈罗德·埃德温·赫斯特是一位杰出的英国水文学家, 他曾参与尼罗河大坝工程。在施工过程中, 赫斯特需要评估水位的波动。最初, 他假设水流是随机注入的随机过程。然而, 赫斯特在研究了尼罗河九个世纪的洪水记录后总结出：高于平均水平的洪灾随后会有更强劲的洪灾。之后这个过程改变其方向, 低于平均水平的洪灾会紧随更弱的洪灾。显然这些是属于非周期性的轮回。

<font size=3.5>赫斯特的统计模型基于阿尔伯特·爱因斯坦关于布朗运动的粒子的随机散布模型。这个理论背后的思路是, 粒子走过的距离 $(R)$ 与时间 $(T)$ 的平方根成比例增加,即 $R=T^{0.5}$。在此研究过程中赫斯特扩展了爱因斯坦的方程, 将其转化为更常用的形式:

<font size=3.5>$$(R/S)_n=C\times n^H$$

<font size=3.5>此处 $c$ 是一个常数。

>### b. Hurst 指数度量粗糙度

```python
def weierstrass_function_H(x,Hurst_exponent,b,c,d):
    """
    计算随机 Weierstrass 函数，生成fBM数据
    
    参数：
    x - 时间序列
    Hurst_exponent - Hurst 指数
    b - Weierstrass 函数的类周期控制参量
    c,d - Weierstrass 函数的随机参量
    
    返回：
    对应x的Weierstrass 函数值
    
    """
    s = 0
    for k in range(len(c)):
        s += c[k] * np.power(b,-(k+1) * Hurst_exponent) * np.sin( np.power(b,k+1) * x + d[k])
    return s
```

>### c. Hurst 指数的估算方法

<font size=3.5>展现时间序列数据中长记忆性的最知名的方法之一是 Mandelbrot 提出的 **重标度涨落极差分析**（R/S），通过去除时间序列子序列中的线性趋势引申的 **去趋势波动分析**（DFA）是 Hurst 估计中的另一种方法。在 Hurst 估计中，最有名的是测量时间序列中增量分布的一阶矩得到的 **广义 Hurst 指数**（GHE）。对 Hurst 指数的估算在近几年间仍发挥着磅礴生命力，提出了基于三个不同时间序列的样本构成 **三角形总面积** 的新方法(TTA)

<font size=3.5>$R/S$、$DFA$、$GHE$和$TTA$有各自的优势和劣势，为了测量一个给定时间序列的 Hurst 指数，我们必须考虑其数据长度或样本数量、时间序列中的噪声水平以及对于实时应用而言更重要的计算所需时间，并根据不同因素选取不同计算方法。在本文中我们主要展示了经典R/S分析和两种针对特定情况下的修正方法。

## 3. *分形维数（Fractal Dimension）*

>### a. 分形维数的计算方法

<font size=3.5>**几何直方图法**：通过统计数据在不同尺度下的分布情况，评估分形的维数。在这种方法中，首先将数据划分为许多等大的区域，然后使用直方图统计每个区域中的数据点数。随着尺度的增加，分布的复杂程度也会随之增加。分形维数是与尺度相关的直方图的趋势线的斜率。

<font size=3.5>**分形直方图法**：使用不同的尺度来构建分形直方图，以评估数据的分形维数。该方法与几何直方图法类似，但它考虑了分形特征的空间关系。

<font size=3.5>**多分形维数方法**（MFD）：使用一组不同的分裂比率对数据进行分析，使用对数对数图评估结果。在 MFD 方法中，每个分裂比率都被用来对数据进行分析，并使用对数对数图评估结果。分形维数是所有分裂比率的趋势线的斜率的平均值。

<font size=3.5>**分形维数算法**：使用特定的数学算法，如灰度共生矩阵（GLCM），模糊熵（FE）和分形复杂度（FC）等来评估分形维数。这些算法考虑了分形图像中的空间和灰度关系，以确定其复杂程度。

<font size=3.5>这些方法的精确度和适用性取决于分形数据的特征和复杂程度。因此，在选择计算分形维数的方法时，应考虑数据的特征和需求。


>### b. 计盒维数

<font size=3.5>在分形几何中, **[计盒维数](https://baike.baidu.com/item/%E8%AE%A1%E7%9B%92%E7%BB%B4%E6%95%B0/22781986)**也称为 **盒维数** 、*闵可夫斯基维数*，是一种测量距离空间(X,d)（特别是豪斯多夫空间）比如欧氏空间R中分形维数的计算方法。要计算分形S的维数，你可以想象一下把这个分形放在一个均匀分割的网格上，数一数最小需要几个格子来覆盖这个分形。通过对网格的逐步精化，查看所需覆盖数目的变化，从而计算出计盒维数。
    
<font size=3.5>假设当格子的边长是$r$时，总共把空间分成N个格子，那么计盒维数就是：$$ D = \lim_{n \to 0} \frac{\text{log}(N_{r})}{\text{log}(1/r)} $$

<font size=3.5>当极限不收敛时，我们必须指出 **顶盒维数** 或 **底盒维数** ，或者说，计盒维数仅在和顶盒维数与底盒维数相等时才是有定义的。顶盒维数也称为能量维数、科莫格洛夫维数、科莫格洛夫容积，或者闵可夫斯基上界维数，类似的可定义闵可夫斯基下界维数。</font>

>### c. 计盒维数的分格展示

```python

def BOX(box_N): # MAX 100,000 得二维矩阵
    """
    生成用于计算计盒维数二维矩阵
    
    参数：
    box_N - 方阵边长
    
    返回：
    box_B - box_N*box_N的方阵
    
    """
    box_B=np.zeros((box_N+1,box_N+1),dtype=np.bool_)
    return box_B
```

```python
#
def Determine_W(step_N,Hurst):  
    """
    生成W函数轨迹点数据以及布尔判据列
    
    参数：
    step_N - 计算步数
    Hurst - 对应Hurst值
    
    返回：
    XY0 - X，Y，布尔三行矩阵
    
    """
    x0=np.linspace(0,1,step_N)
    y0=np.zeros((2,step_N))
    y0[0]=weierstrass_function(x0,Hurst,1.2)
    XY0=np.vstack([x0, y0])

    return XY0

def Determine_W_show(step_N,box_N,Hurst):# 点阵图
    """
    生成W函数轨迹点的分格展示
    
    参数：
    step_N - 计算步数
    Hurst - 对应Hurst值
    box_N - 方阵边长
    
    返回：
    绘图
    
    """    
    Plot_black=Determine_W(step_N,Hurst)
    Plot_black[1]=Plot_black[1]/(np.max(Plot_black[1])-np.min(Plot_black[1]))
    Zplt(Plot_black[0],Plot_black[1],La="分数布朗运动",Al=0.6)
    plt.title("计盒维数的分格展示")
    BOX_k=BOX(box_N)
    
# 确定格数
    B_s=1/box_N
    Bx=np.divide(Plot_black[0],B_s)
    Bx=np.ceil(Bx)
    By=np.divide(Plot_black[1],B_s)
    By=np.ceil(By)
    px=[]
    py=[]

# 格内标识
    
    for i in range(step_N):
        if BOX_k[int(Bx[i]),int(By[i])]==False:
            BOX_k[int(Bx[i]),int(By[i])] = True
            px.append((Bx[i]-0.5)/box_N)
            py.append((By[i]-0.5)/box_N)
  
    Zplt(px,py,Type="S",Ms=5000/box_N,Mm="s",Mc="r",Al=0.4,Size=True,La="分格")
    
    
widgets.interact(
    Determine_W_show,
    box_N=widgets.IntSlider(min=100, max=5000, step=100, value=500,description='N(边等分):'),
    step_N=widgets.IntSlider(min=10000, max=100000, step=1000, value=30000,description='M(行走次数):'),
    Hurst=widgets.FloatSlider(min=0, max=1, step=0.0001, readout_format='.4f',value=0.5,description='H(赫斯特):'),
)
```

# 【 三 】 随机Weierstrass函数的类周期性

<font size=3.5>适当控制随机Weierstrass函数参数b和时间范围的取值，可观察到较为明显的类周期现象。

```python
#
random.seed()
c=np.random.standard_normal(100)
d=np.random.uniform(0,2*np.pi,size=100)
b=np.random.uniform(1,7)
x = np.linspace(0,5,1000)

hurst_exponent = 0.5

y= weierstrass_function_H(x,hurst_exponent,b,c,d)
plt.figure(figsize=[10,5])
plt.plot(x,y,linewidth=1.5,label="原数据")

C=c
for i in range(len(c)):
    C[i]=C[i]*np.power(b,-(i+1)*hurst_exponent)
where=np.argmax(np.abs(C))
y3=2*C[where]*np.power(b,-(where+1)*hurst_exponent) * np.sin(np.power(b,where+1) * x+d[where])
plt.plot(x,y3,linewidth=2,label="最大振幅")
y4=y-y3
plt.plot(x,y4,linewidth=1,label="剩余噪声")
plt.title("随机Weierstrass函数类周期性的主要贡献项")

plt.legend(loc='upper right')
print("原数据标准差为",np.std(y),",减去最大幅值项即第"+str((where+1))+"项后，标准差下降到",np.std(y4),"，即下降"+str(int((np.std(y)-np.std(y4))/np.std(y)*100))+"%"\
    ,"，该变化体现随机Weierstrass函数的类周期性主要由最大振幅项提供。")
plt.show()
```

# 【 四 】 R/S分析估计Hurst指数

>### a. R/S分析——重标度涨落极差

<font size=3.5>**1**. 输入数据为长度为 $M$ 的时间序列数据。将其取对数并做差分，变成长度为 $N=M-1$ 的对数差分序列（注意该时间序列数据均为正数）：

<font size=3.5>$$N_i=\log(M_{i+1}/M_i),i=1,2,\cdots,M-1$$

<font size=3.5>应注意， $R/S$ 不能直接应用于分数布朗运动之中，应取其一阶差分
    
<font size=3.5>**2**. 将长度为 $N$ 的对数收益率序列等分为 $A$ 个子集，每个子集的长度为 $N/A$ 。计算每个子集的均值，记为 $e_a,a=1,2,\cdots,A$。在每个子集 $a$ 内，逐一计算前 $k$ 个点($k=1,2,\cdots,n$)相对该子集均值 $e_a$ 的累积离差：
$$X_{k,a}=\sum_{i=1}^k(N_{i,a}-e_a),k=1,2,\cdots,n$$

<font size=3.5>**3**. 计算每个子集 $a$ 内对数增长率序列的波动范围 $R_a$ ，它等于累积离差最大值和最小值的差值：

<font size=3.5>$$R_a=\text{max}(X_{k,a})-\text{min}(X_{k,a}),1\leq k\leq n$$

<font size=3.5>**4**. 计算每个子集 $a$ 内对数增长率序列的标准差 $S_a$ ，对每个子集 $a$ 内，使用其标准差 $S_a$ 对其波动范围 $R_a$ 进行标准化，得到重标极差 $R_a/S_a$ 。

<font size=3.5>**5**. 从第二步开始，对于选取的长度 $n$ ，我们一共有 $A$ 个子集，因此有 $A$ 个重标极差。取它们的均值作为该原始对数价格序列在长度为 $n$ 的时间跨度上的重标极差，记为 $(R/S)_n$：
$$\begin{aligned}(R/S)_n&=\frac{1}{A}\sum_{a=1}^A\frac{R_a}{S_a}\end{aligned}$$

<font size=3.5>**6**. 增大 $n$ 的取值，并重复前六步，得到不同长度 $n$ 的时间跨度上对数序列的重标极差 $(R/S)_n$ 。

<font size=3.5>**7**. Hurst 指数 $H$ 是描述 $(R/S)_n$ 和 $n^H$ 的正比关系，即：

<font size=3.5>$$(R/S)_n=C\times n^H$$

<font size=3.5>**8**. 对 $(R/S)_n$ 和 $n^H$ 进行双对数回归，即使用 $\text{log}(n)$ 对 $\text{log}((R/S)_n)$
 进行线性回归。回归方程的截距就是上面关系中的常数 $C$ ，而斜率就是 $Hurst$ 指数 
 

>### b.经典R/S分析法

```python
def hurst_exponent_estimation_classics(X,real_h):
    """
    经典方法估算 Hurst 指数
    
    参数：
    x - 时间序列数据
    step - 步长
    
    返回：
    估计 Hurst 指数过程中绘图
    
    """    
    #（对数）差分

    X=np.diff(X)
    
    #取子序列
    R_S = []
    n=int(len(X))
    #子序列长度
    N_step = np.arange(10,int(n/2))
#     number=np.array([2**i for i in range(2,10)])
#     N_step = np.array([int(i) for i in number])
    for step in N_step:
        N = len(X)
        N_del=0
        while (N-N_del)%step != 0:
            N_del+=1
        #不等分时抛弃后面部分数据。
        x=X[:N-N_del]
        #子序列数
        A=int((N-N_del)/step)
        subset_A = np.split(x,A ,axis = 0)

        # 每个子序列的均值
        subset_mean = np.zeros(A)
        # 每个子序列的标准差
        subset_S = np.zeros(A)
        # 每个子序列累计离差的极差
        subset_R = np.zeros(A)
        # 每个子序列的累计离差
        subset_deviation = np.zeros([A,N])
        subset_R_S = np.zeros(A)
        #求累计离差和累计离差的极差
        A_i=0
        for a in subset_A:
            subset_mean[A_i] = np.mean(a)
            subset_S[A_i] = np.std(a,ddof=1) 
            d_i = 0
            #子序列累计离差的极差
            sum_deviation = 0
            for a_i in a:
                sum_deviation = sum_deviation + (a_i-subset_mean[A_i])
                subset_deviation[A_i,d_i]=sum_deviation
                d_i+=1
            deviation_max = np.max(subset_deviation[A_i])
            deviation_min = np.min(subset_deviation[A_i])
            subset_R[A_i] =deviation_max-deviation_min
            #计算重标极差
            subset_R_S [A_i] = subset_R[A_i]/subset_S[A_i]
            A_i+=1
        R_S.append(np.sum(subset_R_S)/A)
    
    H, B = np.polyfit(np.log(N_step),np.log(R_S), 1)
    
    X_show=np.linspace(np.log(np.min(N_step)),np.log(np.max(N_step)),100)
    
    plt.figure(figsize=(9,4))
    plt.xlabel("$log(n)$",size=12)
    plt.ylabel("log(R/S)",size=12)
    plt.title("经典R/S法估算 Hurst 指数",size=16)
    
    plt.scatter(np.log(N_step),np.log(R_S),s=1,label="子序列不同分割方式")
    plt.plot(X_show,X_show*H+B,"g",label=f"估计：ln(R/S)={int(H*1000)/1000} ln(n)+{int(B*1000)/1000}")
    plt.plot(X_show,X_show*real_h+B,'r',label=f"理论：ln(R/S)={int(real_h*1000)/1000}ln(n)+{int(B*1000)/1000}")
    print("实际：",real_h," 估计：",H," 误差："+str(int((-real_h+H)/real_h*10000)/100)+"%")
    plt.legend(loc="best")
    plt.show()
    return ()
```

>### c. R/S-AL修正

<font size=3.5>对于较小的n，白噪声的R/S统计分析与理论上hurst=0.5的数据计算所得斜率有显著的偏差。因此通常取近似：

<font size=3.5>$$\mathbf{E}(R / S)_{n}=\left\{\begin{array}{ll}
\frac{n-\frac{1}{2}}{n} \frac{\Gamma\left(\frac{n-1}{2}\right)}{\sqrt{\pi} \Gamma\left(\frac{n}{2}\right)} \sum_{i=1}^{n-1} \sqrt{\frac{n-i}{i}} & \text { for } \quad n \leq 340 \\
\frac{n-\frac{1}{2}}{n} \frac{1}{\sqrt{n \frac{\pi}{2}}} \sum_{i=1}^{n-1} \sqrt{\frac{n-i}{i}} & \text { for } \quad n>340
\end{array}\right.$$

<font size=3.5>白噪声的赫斯特指数 H 计算为0.5加上 $(R/S)_n - E(R/S)_n$ 的拟合斜率。统计结果用 R/S-A L表示。

```python
def Er(n):
    if n<=340:
        return (n-1/2)/n*math.gamma((n-1)/2)/np.sqrt(np.pi)/math.gamma(n/2)*np.sum(np.sqrt([(n-i)/i for i in range(1,n)]) )
    else:
        return (n-1/2)/n*np.sum(np.sqrt([(n-i)/i for i in range(1,n)]) )/np.sqrt(n*np.pi/2)
```

```python
def hurst_exponent_estimation_Hn(X,real_h):
    """
     AL纠正方法估算 Hurst 指数
    
    参数：
    x - 时间序列数据
    step - 步长
    
    返回：
    估计 Hurst 指数过程中绘图
    
    """    
    #（对数）差分

    X=np.diff(X)
    
    #取子序列
    R_S = []
    logHn=[]
    n=int(len(X))
    #子序列长度
    N_step = np.arange(10,int(n/2))
    for step in N_step:
        N = len(X)
        N_del=0
        while (N-N_del)%step != 0:
            N_del+=1
        #不等分时抛弃后面部分数据。
        x=X[:N-N_del]
        #子序列数
        A=int((N-N_del)/step)
        subset_A = np.split(x,A ,axis = 0)

        # 每个子序列的均值
        subset_mean = np.zeros(A)
        # 每个子序列的标准差
        subset_S = np.zeros(A)
        # 每个子序列累计离差的极差
        subset_R = np.zeros(A)
        # 每个子序列的累计离差
        subset_deviation = np.zeros([A,N])
        subset_R_S = np.zeros(A)
        #求累计离差和累计离差的极差
        A_i=0
        for a in subset_A:
            subset_mean[A_i] = np.mean(a)
            subset_S[A_i] = np.std(a,ddof=1) 
            d_i = 0
            #子序列累计离差的极差
            sum_deviation = 0
            for a_i in a:
                sum_deviation = sum_deviation + (a_i-subset_mean[A_i])
                subset_deviation[A_i,d_i]=sum_deviation
                d_i+=1
            deviation_max = np.max(subset_deviation[A_i])
            deviation_min = np.min(subset_deviation[A_i])
            subset_R[A_i] =deviation_max-deviation_min
            #计算重标极差
            subset_R_S [A_i] = subset_R[A_i]/subset_S[A_i]
            A_i+=1
        R_S.append(np.sum(subset_R_S)/A)
        logHn.append(np.log(np.sum(subset_R_S)/A)-np.log(Er(step))+np.log(step)/2)
    
    H, B = np.polyfit(np.log(N_step),np.log(R_S), 1)

    H_Hn,B_Hn = np.polyfit(np.log(N_step),logHn, 1)

    X_show=np.linspace(np.log(np.min(N_step)),np.log(np.max(N_step)),100)
    
    plt.figure(figsize=(9,4))
    plt.xlabel("$log(n)$",size=12)
    plt.ylabel("log(R/S)",size=12)
    plt.title("R/S_Hn法估算 Hurst 指数",size=16)
    
    plt.scatter(np.log(N_step),np.log(R_S),s=1,label="子序列不同分割方式")
    plt.scatter(np.log(N_step),logHn,s=1,label="Hn下子序列不同分割方式")
    plt.plot(X_show,X_show*H+B,"g",label=f"经典估计：ln(R/S)={int(H*1000)/1000} ln(n)+{int(B*1000)/1000}")
    plt.plot(X_show,X_show*H_Hn+B_Hn,"c",label=f"Hn纠正：ln(R/S)={int(H_Hn*1000)/1000} ln(n)+{int(B_Hn*1000)/1000}")
    plt.plot(X_show,X_show*real_h+B,'r',label=f"理论：ln(R/S)={int(real_h*1000)/1000}ln(n)+{int(B*1000)/1000}")
    print("实际：",real_h," 估计：",H_Hn," 误差："+str(int((-real_h+H_Hn)/real_h*10000)/100)+"%")
    plt.legend(loc="best")
    plt.show()
    return ()
```

# 【 五 】 fBM 数据的分形维数

>### a. 计盒维数法

```python
def Er(n):
    if n<=340:
        return (n-1/2)/n*math.gamma((n-1)/2)/np.sqrt(np.pi)/math.gamma(n/2)*np.sum(np.sqrt([(n-i)/i for i in range(1,n)]) )
    else:
        return (n-1/2)/n*np.sum(np.sqrt([(n-i)/i for i in range(1,n)]) )/np.sqrt(n*np.pi/2)
```

```python
def Determine_W_N(Hurst,step_N,box_S):# 得保留格数
    """
    计算当前条件下覆盖该分形需要最少格子数
    
    参数：
    step_N - 时间序列数据量
    Hurst - Hurst 指数
    box_S - 盒子边长
    
    返回：
    sum - 对应条件下覆盖该分形需要最少格子数
    
    """    
#     Source=np.diff(Determine_W(step_N,Hurst))
    Source=Determine_W(step_N,Hurst)
    
# 确定格数
    #盒子边长
    B_s = box_S
    Bx=np.divide(Source[0],B_s)
    Bx=np.ceil(Bx)
    By=np.divide(Source[1],B_s)
    By=np.ceil(By)
    Bx_min=int(np.min(Bx))
    By_min=int(np.min(By))
    box_Nx=np.max(Bx)-Bx_min
    box_Ny=np.max(By)-By_min
    box_N=int(max(box_Nx,box_Ny))
#     print(box_S,box_N,box_Nx,box_Ny)
    BOX_k=BOX(box_N+1)

# 格内标识
    num=0
    for i in range(step_N):
#         if BOX_k[int(Bx[i-1])-Bx_min,int(By[i-1])-By_min]==False:
#             BOX_k[int(Bx[i-1])-Bx_min,int(By[i-1])-By_min] = True
#             num=num+1
        if BOX_k[int(Bx[i])-Bx_min,int(By[i])-By_min]==False:
            BOX_k[int(Bx[i])-Bx_min,int(By[i])-By_min] = True
            num=num+1
    
    return num


def func(x,k,b):
    return(k*x+b)
```

```python
def weierstrass_function(x,Hurst_exponent,b):
    """
    计算随机 Weierstrass 函数，生成fBM数据
    
    参数：
    x - 时间序列
    Hurst_exponent - Hurst 指数
    b - Weierstrass 函数的类周期控制参量
    
    返回：
    对应x的Weierstrass 函数值
    
    """
    s = 0
    random.seed()
    #取前100项
    c=np.random.standard_normal(1000)
    d=np.random.uniform(0,2*np.pi,size=1000)
    for k in range(1000):

        s += c[k] * np.power(b,-k * Hurst_exponent) * np.sin(  np.power(b,k) * x + d[k])
    return s
```

```python

def D_main2_show(N_2sq,W_N,Hurst):
    
    """
    对数拟合得该分形分形维数
    
    参数：
    x - 时间序列
    Hurst_exponent - Hurst 指数
    b - Weierstrass 函数的类周期控制参量
    
    返回：
    对应x的Weierstrass 函数值
    
    """
    x=[]
    y=[]

    for i in range(1,N_2sq+1):
            y.append(np.log(Determine_W_N(Hurst,W_N,1/(2**i))))
            x.append(np.log(2**i))     
            
    Zplt(x,y,Size=(9,5),Mm="o",Mc="r",Ms=5,La="计盒维数",Tn=f"计盒维数随盒边长变化,边等分{2**N_2sq}份",Lx='$ln(1/r)$',Ly="$ln{N_r}$") 

    popt, pcov = curve_fit(func,x,y)
#     print(popt, pcov)
    x0=np.linspace(min(x),max(x),100)
    Zplt(x0,func(x0,popt[0],popt[1]),La=f"拟合曲线,D={int(popt[0]*10000)/10000}",Lc="b",Size=True)  
    Zplt(x0,func(x0,2-Hurst,popt[1]),La=f"标准曲线,D={int((2-Hurst)*10000)/10000}",Lc="g",Size=True)
    
widgets.interact_manual(
    D_main2_show,
    N_2sq=widgets.IntSlider(min=3, max=14, step=1, value=9,description='N(分格级数):'),
    W_N=widgets.IntSlider(min=10000, max=1000000, step=100, value=10000,description='M(行走次数):'),
    Hurst=widgets.FloatSlider(min=0, max=1, step=0.000001, value=0.5,description='H:'),
    计盒维数=False,
    LOG=False
)
```

# 【 六 】 Hurst指数与分形维数的数量关系

>### a.  H + D = 2的直观理解

<font size=3.5>在时间序列中，分形维数可视为：$$ D = \lim_{n \to 0} \frac{\text{log}(N_{r})}{\text{log}(1/r)} = \lim_{n \to 0} \frac{\text{log}(N_{h})}{\text{log}(1/h)} $$

<font size=3.5>其中 $h$ 是时间增量,即计盒维数的盒边长 $r$，$N_h$ 是与时间序列图形相交的正方形的数量。假设时间序列函数$f$是连续的，在每个子区间 $[kh,(k + 1)h]$ 内将有一个大约 $R/h$ 个立方体的平均值。由于有子区间长 $h$，总长为1，共有 $1/h$ 个子区间。所以总共大约有 $\frac{R_h}{h^2}$个正方形，因此得出 $N_h =R_h*h^2$

<font size=3.5>由文献中对 $H$ 的形象理解：
$$H^{'}=\lim\limits_{h\to0}\dfrac{\log R_{h}}{\log h}=\lim\limits_{h\to0}\dfrac{\log N_{h}+\log h^{2}}{\log h}=\lim\limits_{h\to0}\dfrac{\log N_{h}}{\log h}+2=2-d$$

>### b.  R/S_Hn与修正的计盒维数

```python
def D_estimation(W_N,Hurst):
    
    """
    计算时间序列数据的计盒维数
    
    参数：
    W_N - 时间序列数据量
    Hurst - Hurst 指数
    
    返回：
    对应的分形维数
    
    """
    x=[]
    y=[]
    Y=[]
    for i in range(1,10):
        y.append(np.log(Determine_W_N(Hurst,W_N,1/(2**i))))
        x.append(np.log(2**i))
    i = 4
    while (y[i]-2*y[i-1]+y[i-2])>=0 and i<=9:
        i+=1
    popt, pcov = curve_fit(func,x[:i],y[:i])
#     popt, pcov = curve_fit(func,x,y)
    return popt[0]
```

# 【 七 】 结论与展望

<font size=4> <div style="width:3.5">本文档着重讨论了重标极差分析方法及其修正的有效性。借助分数布朗运动进一步证实了Hurst指数与分形盒维数之和为2的关系。设定不同赫斯特指数通过随机魏尔斯特拉斯函数模拟出一系列分数高斯噪声时间序列数据，在分析过程中通过给定H下各随机项取随机值演示了随着H值增大，图像逐渐趋于平滑的整个过程。在改变b时研究其类周期性，体现随机Weierstrass函数的类周期性主要由最大振幅项提供。H指数的计算在R/S分析中是通过双对数回归分析方法得到，其估计值的显著性检验同n的取值大小密切相关。n取值较大时，H指数估计值的显著性检验效果明显，n取值较小时，检验效果较差。在选择计算分形维数的方法时，考虑数据的特征和需求采取计盒维数法。在误差允许范围之内，Hurst指数与分形维数的数量关系为：H=2-D。
