在数学上，对于一个在 ${\color{Blue}{实数}}$ 或 ${\color{Blue}{复数}}$ 复数a邻域上，${\color{Blue}{以实数作为变量}}$ 或 ${\color{Blue}{以复数作为变量}}$ 的 ${\color{Blue}{函数}}$ ，并且是 ${\color{Blue}{无穷可微的函数}}$ $f(x)$ ,它的 **泰 勒 级 数** 是以下这种形式的 ${\color{Blue}{幂级数}}$ ：

$\sum_{n=0}^{+\infty} \frac{f^{(n)}(a)}{n!}{(x-a)^n}$

这里，${n!}$ 表示n的 ${\color{Blue}{阶乘}}$ ，而 ${f^{(n)}(a)}$ 表示函数 $f$ 在点a处的n阶 ${\color{Blue}{导数}}$ 。如果a=0，也可以把这个级数称为 **麦 克 劳 林 级 数**。


### 1.泰勒公式定义
$$f(x) = \frac{f(x_0)}{0!} + \frac{f'(x_0)}{1!}(x-x_0) +\frac{f''(x_0)}{2!}(x-x_0)^2 + \cdots + \frac{f^{(n)}(x_0)}{n!}(x-x_0)^{n} +  R_n(x) =\sum\limits_{n=0}^{\infty}{\frac{f^{(n)}(x_0)}{n!}(x-x_0)^{n}} $$
 
 
 $$ R_n(x) = o[(x-x_0)^{n}] = \frac{f^{(n+1)}(\zeta)}{(n+1)!}(x-x_0)^{n+1}$$ 
 其中  $\zeta\in[x_0,x] $

### 2.常见函数的TAYLOR展开式(麦克劳林)

$$  e^x = 1+x+\frac{x^2}{2!}+\cdots+\frac{x^{n}}{n!}+o(x^{n})$$

$$  \frac{1}{1+x}=1-x+x^2+\cdots+(-1)^{n}x^{n}+o(x^{n})$$

$$ ln(1+x) =x-\frac{x^2}{2}+\frac{x^3}{3}+\cdots+(-1)^{n-1}\frac{x^n}{n}+o(x^{n})$$

$$ sin x=x-\frac{x^3}{3!}+\frac{x^5}{5!}+\cdots+(-1)^{k-1}\frac{x^{2k-1}}{(2k-1)!}-o(x^{2k-1})$$

$$ cos x=1-\frac{x^2}{2!}+\frac{x^4}{4!}+\cdots+(-1)^{k}\frac{x^{2k}}{(2k)!}-o(x^{2k})$$
